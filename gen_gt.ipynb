```python
T1_CLASS_NAMES = [
    "aeroplane","bicycle","bird","boat","bus","car",
    "cat","cow","dog","horse","motorbike","sheep","train",
    "elephant","bear","zebra","giraffe","truck","person"
]

T2_CLASS_NAMES = [
    "traffic light","fire hydrant","stop sign",
    "parking meter","bench","chair","diningtable",
    "pottedplant","backpack","umbrella","handbag",
    "tie","suitcase","microwave","oven","toaster","sink",
    "refrigerator","bed","toilet","sofa"
]

T3_CLASS_NAMES = [
    "frisbee","skis","snowboard","sports ball",
    "kite","baseball bat","baseball glove","skateboard",
    "surfboard","tennis racket","banana","apple","sandwich",
    "orange","broccoli","carrot","hot dog","pizza","donut","cake"
]

T4_CLASS_NAMES = [
    "laptop","mouse","remote","keyboard","cell phone","book",
    "clock","vase","scissors","teddy bear","hair drier","toothbrush",
    "wine glass","cup","fork","knife","spoon","bowl","tvmonitor","bottle"
]
```

```python
import os
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
def parse_xml(directory, output_dir):
    data = []

    for file in os.listdir(directory):
        if file.endswith('.xml'):
            tree = ET.parse(os.path.join(directory, file))
            root = tree.getroot()

            filename = root.find('filename').text.split('.jpg')[0]
            if len(filename)==12:
                for obj in root.findall('object'):
                    label = obj.find('name').text
                    if label=="airplane": label="aeroplane"
                    if label=="motorcycle": label="motorbike"
                        
                    if label not in T1_CLASS_NAMES:
                        label = 'unknown'
                    
                    bbox = obj.find('bndbox')
                    x_min = bbox.find('xmin').text
                    y_min = bbox.find('ymin').text
                    x_max = bbox.find('xmax').text
                    y_max = bbox.find('ymax').text
    
                    data.append([filename, x_min, y_min, x_max, y_max, label, 1.0])

    # Convert the data to DataFrame and save as CSV
    columns = ["img_name", "x_min", "y_min", "x_max", "y_max", "label", "prob"]
    df = pd.DataFrame(data, columns=columns)
    csv_path = os.path.join(output_dir)
    df.to_csv(csv_path, index=False)

    return data
```

```python
directory = "/training_data_2/yuetian/OWOD/Annotations"
output = "./csv/bbox_data_scaled_gt_t1.csv"
parsed_data = parse_xml(directory, output)
parsed_data
```
